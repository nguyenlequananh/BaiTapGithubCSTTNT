```python
import numpy as np
import matplotlib.pyplot as plt

# S lng hnh trn
N = 50

# Sinh ta  ngu nhin
x = np.random.rand(N)
y = np.random.rand(N)

# Sinh mu sc ngu nhin
colors = np.random.rand(N)

# Sinh bn knh v tnh din tch hnh trn
radii = np.random.rand(N)
area = np.pi * (radii * 15) ** 2  # Nhn 15  tng kch thc hnh trn

# Hin th  th
plt.figure(figsize=(6, 6))
plt.scatter(x, y, s=area, c=colors, alpha=0.6, edgecolors="black")
plt.xlabel("Trc X")
plt.ylabel("Trc Y")
plt.title("Biu  50 hnh trn ngu nhin")
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()
```
