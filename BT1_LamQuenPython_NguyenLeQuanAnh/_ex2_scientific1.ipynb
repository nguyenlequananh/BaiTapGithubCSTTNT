```python
import numpy as np
import matplotlib.pyplot as plt

# Số lượng hình tròn
N = 50

# Sinh tọa độ ngẫu nhiên
x = np.random.rand(N)
y = np.random.rand(N)

# Sinh màu sắc ngẫu nhiên
colors = np.random.rand(N)

# Sinh bán kính và tính diện tích hình tròn
radii = np.random.rand(N)
area = np.pi * (radii * 15) ** 2  # Nhân 15 để tăng kích thước hình tròn

# Hiển thị đồ thị
plt.figure(figsize=(6, 6))
plt.scatter(x, y, s=area, c=colors, alpha=0.6, edgecolors="black")
plt.xlabel("Trục X")
plt.ylabel("Trục Y")
plt.title("Biểu đồ 50 hình tròn ngẫu nhiên")
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()
```
